```python
import pandas as pd
import json
from shapely.geometry import Polygon
```

```python
json_file='us-state-boundaries.json'
def m_func(state:str,df:pd.DataFrame)->list:
    """Возвращает список штатов соседей"""
    neighbors=[]
    for i in df['State'][df['State']!=state]:
        if Polygon(df[df['State']==state]['Polygon'].values[0]).intersects(Polygon(df[df['State']==i]['Polygon'].values[0])):
            neighbors.append(i)
    return neighbors
def make_dataframe_with_neighbors(json_file:str)->pd.DataFrame:
    """Формирует датафрейм из следующих колонок:
    State: Штат,
    Polygon: Координаты штата(полигон координат),
    Neighbors: Соседи штата
    """
    df=pd.DataFrame(columns=['State','Polygon'])
    num_states=len(json.load(open(json_file, encoding='utf-8')))
    for i in range(num_states):
        if json.load(open(json_file, encoding='utf-8'))[i]['fields']['st_asgeojson']['type']=='MultiPolygon':
            df=df.append({'State': json.load(open(json_file, encoding='utf-8'))[i]['fields']['basename'],
                   'Polygon': json.load(open(json_file, encoding='utf-8'))[i]['fields']['st_asgeojson']['coordinates'][0][0]},ignore_index=True)
        else:
            df=df.append({'State': json.load(open(json_file, encoding='utf-8'))[i]['fields']['basename'],
                   'Polygon': json.load(open(json_file, encoding='utf-8'))[i]['fields']['st_asgeojson']['coordinates'][0]},ignore_index=True)

    df['Neighbors']=df['State'].apply(lambda x:m_func(x,df))
    return df
```

```python
"""Проверяю работу"""
make_dataframe_with_neighbors(json_file)
```
